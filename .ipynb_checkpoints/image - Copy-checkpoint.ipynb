### Load Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

### Prepare/collect data

```python
import os

path = os.listdir('brain_tumor/Training/')
classes = {'no_tumor':0, 'pituitary_tumor':1, 'meningioma_tumor':1}
```

```python
import cv2
xtrain = []
ytrain = []
for cls in classes:
    pth = 'brain_tumor/Training/'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j, 0)
        img = cv2.resize(img, (200,200))
        xtrain.append(img)
        ytrain.append(classes[cls])
xtest = []
ytest = []
for cls in classes:
    pth = 'brain_tumor/Testing/'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j, 0)
        img = cv2.resize(img, (200,200))
        xtest.append(img)
        ytest.append(classes[cls])
```

```python
xtrain = np.array(xtrain)
ytrain = np.array(ytrain)

xtest = np.array(xtest)
ytest = np.array(ytest)

xtrain_updated = xtrain.reshape(len(xtrain), -1)
xtest_updated = xtest.reshape(len(xtest), -1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.fit_transform(xtest)
```

### Feature Selection: PCA

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis  as LDA
```

```python
#print(xtrain.shape, xtest.shape)

pca = PCA()
xtrain = pca.fit_transform(xtrain,ytrain)
xtest = pca.transform(xtest)
```

### Train Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
```

```python
import warnings
warnings.filterwarnings('ignore')

lg = LogisticRegression(C=0.1)
lg.fit(xtrain, ytrain)
```

```python
sv = SVC()
sv.fit(xtrain, ytrain)
```

```python
sgd = SGDClassifier()
sgd.fit(xtrain,ytrain)
```

### Evaluation

```python
print("Training Score:", lg.score(xtrain, ytrain))
print("Testing Score:", lg.score(xtest, ytest))
```

```python
print("Training Score:", sv.score(xtrain, ytrain))
print("Testing Score:", sv.score(xtest, ytest))
```

```python
print("Training Score:", sgd.score(xtrain, ytrain))
print("Testing Score:", sgd.score(xtest, ytest))
```
